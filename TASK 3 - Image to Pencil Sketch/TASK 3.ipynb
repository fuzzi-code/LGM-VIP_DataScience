## Beginner Task 3 : Image to Pencil Sketch with Python

### Importing Modules

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Loading Original Image

```python
img = cv2.imread("sample.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(13,13))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

### Converting Image to GrayScale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(13,13))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

### Inverting the Image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(13,13))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

### Blurring/Smoothing the Image

```python
img_blur = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(12,12))
plt.imshow(img_blur,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```

### Converting your images to pencil sketches

```python
final = cv2.divide(img_gray, 255 - img_blur, scale=255)
plt.figure(figsize=(13,13))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

### Comparing the original and final images

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
